# 4.Preprocess the table from missing, outlier, duplicate, etc.


```python
# detect the features that has nulls
merged_temps.isnull().sum()
```

```python
preprocessedTable = merged_temps.copy()
```

```python
# Fill missing values with the mean based on groups (grouped by dt and Country)
preprocessedTable['AverageTemperature'] = preprocessedTable.groupby(['dt', 'Country'])['AverageTemperature'].transform(lambda x: x.fillna(x.mean()))
preprocessedTable['AverageTemperatureUncertainty'] = preprocessedTable.groupby(['dt', 'Country'])['AverageTemperatureUncertainty'].transform(lambda x: x.fillna(x.mean()))
```

```python
preprocessedTable['AverageTemperature'].isnull().sum()
preprocessedTable['AverageTemperatureUncertainty'].isnull().sum()
```

```python
# Handle remaining nulls with global mean
preprocessedTable['AverageTemperature'] = preprocessedTable.groupby(['dt'])['AverageTemperature'].transform(lambda x: x.fillna(x.mean()))
preprocessedTable['AverageTemperatureUncertainty'] = preprocessedTable.groupby(['dt'])['AverageTemperatureUncertainty'].transform(lambda x: x.fillna(x.mean()))
```

```python
preprocessedTable['AverageTemperature'] = preprocessedTable['AverageTemperature'].transform(lambda x: x.fillna(x.mean()))
preprocessedTable['AverageTemperatureUncertainty'] = preprocessedTable['AverageTemperatureUncertainty'].transform(lambda x: x.fillna(x.mean()))

preprocessedTable.isnull().sum()
```

```python
# print the entities that has null country but does have a city or a state
preprocessedTable[(preprocessedTable['Country'].isnull()) & (preprocessedTable['City'].notnull()) & (preprocessedTable['State'].notnull())]

# it is 0 then we can not predict it, so we just gonna label it as unknown
preprocessedTable['Country'].fillna('unknown', inplace=True)
```

```python
# find entities that has a country  but not a city
preprocessedTable[(preprocessedTable['Country'].notnull()) & (preprocessedTable['City'].isnull()) & (preprocessedTable['State'].isnull())]
```

```python
# find the most repeated city in each country and fill the null cities based on that if not then just fill it with unknown
preprocessedTable['City'] = preprocessedTable.groupby('Country')['City'].transform(lambda x: x.fillna(x.mode().iloc[0] if not x.mode().empty else 'unknown'))
preprocessedTable['City'].isnull().sum()
```

```python
# find the most repeated (mode) in a country and a city then fill null states based on it
preprocessedTable['State'] = preprocessedTable.groupby(['Country', 'City'])['State'].transform(lambda x: x.fillna(x.mode().iloc[0] if not x.mode().empty else 'unknown'))
preprocessedTable['State'].isnull().sum()
```

```python
preprocessedTable.isnull().sum()
```

```python
# I wanted to predict the the latitude and longitude based on the city and the country using geopy
# but unfortuntly the session keeps crashing and no enough RAM is available
# so i am gonna make the null latitude and longitude as 0,0
preprocessedTable['Latitude'].fillna(0, inplace=True)
preprocessedTable['Longitude'].fillna(0, inplace=True)
preprocessedTable.isnull().sum()
```

```python
# Detect duplicates
duplicates = preprocessedTable[preprocessedTable.duplicated()]
print(f"Number of duplicates: {len(duplicates)}")

preprocessedTable.drop_duplicates(inplace=True)

duplicates = preprocessedTable[preprocessedTable.duplicated()]
print(f"Number of duplicates: {len(duplicates)}")
```

```python
preprocessedTable.head()
```

```python
preprocessedTable.columns
```

```python
processedTable = preprocessedTable.copy()
```

```python
# fill the land max is based on the avg temp grouped by the date time
# find the max value based on the avg temp
# where entity state and country and city is not unknown

# Fill 'LandMinTemperature' based on the minimum 'AverageTemperature' for each date
# where 'City', 'State', and 'Country' are not 'unknown'

filteredTable = processedTable[
    (processedTable['City'] != 'unknown') &
    (processedTable['State'] != 'unknown') &
    (processedTable['Country'] != 'unknown')
]

max_values = filteredTable.groupby('dt')['AverageTemperature'].max()
max_uncertainty = filteredTable.groupby('dt')['AverageTemperatureUncertainty'].max()

mean_values = filteredTable.groupby('dt')['AverageTemperature'].mean()
mean_uncertainty = filteredTable.groupby('dt')['AverageTemperatureUncertainty'].mean()

min_values = filteredTable.groupby('dt')['AverageTemperature'].min()
min_uncertainty = filteredTable.groupby('dt')['AverageTemperatureUncertainty'].min()

processedTable['LandMaxTemperature'] = processedTable['dt'].map(max_values)
processedTable['LandMaxTemperatureUncertainty'] = processedTable['dt'].map(max_uncertainty)

processedTable['LandMaxTemperature'] = processedTable['LandMaxTemperature'].fillna(
    processedTable.groupby('dt')['AverageTemperature'].transform('max')
)
processedTable['LandMaxTemperatureUncertainty'] = processedTable['LandMaxTemperatureUncertainty'].fillna(
    processedTable.groupby('dt')['AverageTemperatureUncertainty'].transform('max')
)
```

```python
processedTable['LandMinTemperature'] = processedTable['dt'].map(min_values)
processedTable['LandMinTemperatureUncertainty'] = processedTable['dt'].map(min_uncertainty)

processedTable['LandMinTemperature'] = processedTable['LandMinTemperature'].fillna(
    processedTable.groupby('dt')['AverageTemperature'].transform('min')
)
processedTable['LandMinTemperatureUncertainty'] = processedTable['LandMinTemperatureUncertainty'].fillna(
    processedTable.groupby('dt')['AverageTemperatureUncertainty'].transform('min')
)

processedTable.isnull().sum()
```

```python
# fill land avg
processedTable['LandAverageTemperature'] = processedTable['dt'].map(mean_values)
processedTable['LandAverageTemperatureUncertainty'] = processedTable['dt'].map(mean_uncertainty)

processedTable['LandAverageTemperature'] = processedTable['LandAverageTemperature'].fillna(
    processedTable.groupby('dt')['AverageTemperature'].transform('mean')
)
processedTable['LandAverageTemperatureUncertainty'] = processedTable['LandAverageTemperatureUncertainty'].fillna(
    processedTable.groupby('dt')['AverageTemperatureUncertainty'].transform('mean')
)

processedTable.isnull().sum()
```

```python
# fill the ocean by getting all the unkown and known city and country and state

processedTable['LandAndOceanAverageTemperature'] = processedTable.groupby('dt')['AverageTemperature'].transform(
    lambda x: x.fillna(x.mean() if len(x) > 0 else 0))
processedTable['LandAndOceanAverageTemperatureUncertainty'] = processedTable.groupby('dt')['AverageTemperatureUncertainty'].transform(
    lambda x: x.fillna(x.mean() if len(x) > 0 else 0))
```
